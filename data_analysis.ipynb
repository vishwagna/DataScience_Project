```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#data analysis


# Load the cleaned data from the CSV file
cleaned_data = pd.read_csv('transformed_data.csv')

#analysing participation of nso students in years 2022 and 2023


# Filter NSO students (part of NSO) who participated in 2022
nso_participants_2022 = cleaned_data[(cleaned_data['PART OF NSO OR NOT'] != 0) & (cleaned_data['PARTICIPATION IN 2022'] == 1)]


# Filter NSO students (part of NSO) who participated in 2023
nso_participants_2023 = cleaned_data[(cleaned_data['PART OF NSO OR NOT'] != 0) & (cleaned_data['PARTICIPATION IN 2023'] == 1)]

# Filter NSO students (part of NSO) who participated in both 2022 and 2023
nso_participants_both_years = cleaned_data[
    (cleaned_data['PART OF NSO OR NOT'] != 0) & 
    (cleaned_data['PARTICIPATION IN 2022'] == 1) & 
    (cleaned_data['PARTICIPATION IN 2023'] == 1)
]

# Calculate the percentages
total_nso_students = len(cleaned_data[cleaned_data['PART OF NSO OR NOT'] != 0])
percentage_2022 = (len(nso_participants_2022) / total_nso_students) * 100
percentage_2023 = (len(nso_participants_2023) / total_nso_students) * 100
percentage_both_years = (len(nso_participants_both_years) / total_nso_students) * 100

print("Percentage of nso students who participated in year 2022:",percentage_2022)
print("Percentage of nso students who participated in year 2023:",percentage_2023)
print("Percentage of nso students who participated in both the years:",percentage_both_years)

# Create a bar graph
participation_labels = ['2022', '2023', 'Both Years']
participation_percentages = [percentage_2022, percentage_2023, percentage_both_years]

plt.bar(participation_labels, participation_percentages)
plt.xlabel('Year')
plt.ylabel('Percentage of NSO Students Participating')
plt.title('NSO Student Participation in 2022 and 2023')
plt.show()
```

```python
#analysing participation of nso students based on gender in years 2022 and 2023
cleaned_data = pd.read_csv('transformed_data.csv')

#nso girls participation in year 2022
nso_2022_girls=cleaned_data[(cleaned_data['PART OF NSO OR NOT'] != 0) & (cleaned_data['PARTICIPATION IN 2022'] == 1) & (cleaned_data['GENDER'] == "Female") ]
percentage_girls_2022= (len(nso_2022_girls) / len(nso_participants_2022)) * 100
print("Percentage of nso students(girls) who participated in year 2022:",percentage_girls_2022)


#non nso girls participation in year 2022
participants_2022=cleaned_data[(cleaned_data['PARTICIPATION IN 2022'] == 1)]
nso_2022_girls1=cleaned_data[(cleaned_data['PART OF NSO OR NOT'] ==0) & (cleaned_data['PARTICIPATION IN 2022'] == 1) & (cleaned_data['GENDER'] == "Female") ]
percentage_girls1_2022= (len(nso_2022_girls1) / len(participants_2022)) * 100
print("Percentage of non nso students(girls) who participated in year 2022:",percentage_girls1_2022)


#nso girls participation in year 2023
nso_2023_girls=cleaned_data[(cleaned_data['PART OF NSO OR NOT'] != 0) & (cleaned_data['PARTICIPATION IN 2023'] == 1) & (cleaned_data['GENDER'] == "Female") ]
percentage_girls_2023= (len(nso_2023_girls) / len(nso_participants_2023)) * 100
print("Percentage of nso students(girls) who participated in year 2023:",percentage_girls_2023)


#non nso girls participation in year 2023
participants_2023=cleaned_data[(cleaned_data['PARTICIPATION IN 2023'] == 1)]
nso_2023_girls1=cleaned_data[(cleaned_data['PART OF NSO OR NOT'] ==0) & (cleaned_data['PARTICIPATION IN 2023'] == 1) & (cleaned_data['GENDER'] == "Female") ]
percentage_girls1_2023= (len(nso_2023_girls1) / len(participants_2023)) * 100
print("Percentage of non nso students(girls) who participated in year 2023:",percentage_girls1_2023)


#nso girls participation in both years 
nso_both_girls=cleaned_data[(cleaned_data['PART OF NSO OR NOT'] != 0) & (cleaned_data['PARTICIPATION IN 2022'] == 1) & (cleaned_data['PARTICIPATION IN 2023'] == 1)& (cleaned_data['GENDER'] == "Female") ]
percentage_girls_both= (len(nso_both_girls) / len(nso_participants_both_years)) * 100
print("Percentage of nso students(girls) who participated in both the years:",percentage_girls_both)


#non nso girls participation in both years 
participants_both_years=cleaned_data[(cleaned_data['PARTICIPATION IN 2023'] == 1) & (cleaned_data['PARTICIPATION IN 2022'] == 1)]
nso_both_girls1=cleaned_data[(cleaned_data['PART OF NSO OR NOT'] == 0) & (cleaned_data['PARTICIPATION IN 2022'] == 1) & (cleaned_data['PARTICIPATION IN 2023'] == 1)& (cleaned_data['GENDER'] == "Female") ]
percentage_girls1_both= (len(nso_both_girls1) / len(participants_both_years)) * 100
print("Percentage of non nso students(girls) who participated in both the years:",percentage_girls1_both)


# Create a bar graph
participation_labels = ['2022', '2023', 'Both Years']
participation_percentages = [percentage_girls1_2022, percentage_girls1_2023, percentage_girls1_both]

plt.bar(participation_labels, participation_percentages)
plt.xlabel('Year')
plt.ylabel('Percentage of NON NSO Girl Students Participating')
plt.title('NON NSO Girl Student Participation in 2022 and 2023')
plt.show()


# Create a bar graph
participation_labels = ['2022', '2023', 'Both Years']
participation_percentages = [percentage_girls_2022, percentage_girls_2023, percentage_girls_both]

plt.bar(participation_labels, participation_percentages)
plt.xlabel('Year')
plt.ylabel('Percentage of NSO Girl Students Participating')
plt.title('NSO Girl Student Participation in 2022 and 2023')
plt.show()
```

```python

# Load the cleaned data from the CSV file
cleaned_data = pd.read_csv('transformed_data.csv')
# Create a second set of subplots for participation
fig1, axes1 = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
# Iterate over years and departments for participation
for i, year in enumerate([2022, 2023]):
    for j, department in enumerate(['CSE', 'ECE', 'MECH']):
        ax = axes1[i, j]
        df_year_dept = cleaned_data[(cleaned_data[f'PARTICIPATION IN {year}'] == 1) & (cleaned_data['DEPARTMENT'] == department)]
        sport_counts = df_year_dept['GAME'].value_counts()
        sport_counts.plot(kind='bar', ax=ax)
        ax.set_ylabel("Number of Participants")
        ax.set_title(f"Sports-wise Participation of {department} in {year}")
        ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
df = pd.read_csv('transformed_data.csv')

# Create a dictionary to store the results
results = {"Sports": [], "2022": [], "2023": [], "Both": []}

# Get a list of unique sports
sports = df["GAME"].unique()

# Calculate percentages for each sport
for sport in sports:
    total_participants_2022 = df[df["GAME"] == sport]["PARTICIPATION IN 2022"].sum()
    total_participants_2023 = df[df["GAME"] == sport]["PARTICIPATION IN 2023"].sum()
    total_participants_both = df[(df["GAME"] == sport) & (df["PARTICIPATION IN 2022"] == 1) & (df["PARTICIPATION IN 2023"] == 1)].shape[0]
    
    results["Sports"].append(sport)
    results["2022"].append((total_participants_2022 / len(df[df["PARTICIPATION IN 2022"] == 1])) * 100)
    results["2023"].append((total_participants_2023 / len(df[df["PARTICIPATION IN 2023"] == 1])) * 100)
    results["Both"].append((total_participants_both / len(df[df["PARTICIPATION IN 2022"] == 1])) * 100)

# Create a DataFrame to display the results
results_df = pd.DataFrame(results)

# Print the results
print("Participation percentage in each sport year wise:")
print(results_df)
```

```python
# Load your dataset
data = pd.read_csv('transformed_data.csv')

# Filter students who are winners in either 2022 or 2023
winners_df = data[(data['WINNERS OF 2022'] == 1) | (data['WINNERS OF 2023'] == 1)]

# Save the filtered dataset to a new CSV file
winners_df.to_csv('winners.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('winners.csv')

# Group the data by department and calculate the total winners for each year
winners_by_department = df.groupby('DEPARTMENT')[['WINNERS OF 2022', 'WINNERS OF 2023']].sum()

# Plot a bar chart
winners_by_department.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Winners by Department in 2022 and 2023')
plt.xlabel('Department')
plt.ylabel('Number of Winners')
plt.legend(title='Year', labels=['2022', '2023'])
plt.xticks(rotation=0)  # Keep x-axis labels horizontal

# Show the plot
plt.show()
```
